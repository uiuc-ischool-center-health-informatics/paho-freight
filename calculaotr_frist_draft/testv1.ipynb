```python

print("hello there")

import warnings
warnings.filterwarnings('ignore')


import ipywidgets as widgets
from IPython.display import display, clear_output

file = open("poinappel.jpg", "rb")
image= file.read()

image_headline = widgets.Image(
    value=image,
    format='jpg',
    width='1440'
)

label_headline = widgets.Label(
    value='a delicious fruit',
    style={'description_width': 'initial'}
)

vbox_headline = widgets.VBox([image_headline, label_headline])

poinappel = widgets.ToggleButtons(
    options=['fruit', 'vegetable']
)

name = widgets.Text(placeholder='Your name here')

date = widgets.DatePicker(description='Pick a DATE')


friends = widgets.IntSlider(
    value=0, #default val is u have no friends
    min = 0, 
    max = 1, #no way you have more than one friend 
    style={'description_width': 'initial', 'handle_color': '#16a085'}
)

button_send = widgets.Button(
    description='Tell the poinappel wheter it is a fruit or a vegetable',
    tooltip='Send',
    style={'description_width': 'initial'}
)

output = widgets.Output()

def on_button_clicked(event):
    with output:
        clear_output()
        print("Sent message: ")
        print(f"Dera mr poinappel, this is your favorite {poinappel.value}, {name.value}.")
        print(f"I would love to come over on {date.value} and let you know that you are such a {poinappel.value}, if that's okay with you!")
        if (friends == 0):
            print(f"ALso i have no friends so itll just be you and me! :D")
        else:
            print(f"Also I hop eu don't mind i'm going to be bringing over {friends.value} along with me. merry chirstmas!")

button_send.on_click(on_button_clicked)

vbox_result = widgets.VBox([button_send, output])

text_0 = widgets.HTML(value="<h1>hey mr poinappel! </h1>")
text_1 = widgets.HTML(value="<h2>This is your favorite </h2>")
text_2 = widgets.HTML(value="<h2>for dinner, if that's okay for you! </h2>")
text_3 = widgets.HTML(value="<h2> ALSO if you're don't mind I brought along </h2>")
text_4 = widgets.HTML(value="<h2> friends over for dinner!! yay!!!")
text_5 = widgets.HTML(value="<h2> bababooy </h2>")

vbox_text = widgets.VBox([text_0, text_1, poinappel, name, text_2, date, text_3, text_4, friends, text_5, vbox_result])

page = widgets.HBox([vbox_headline, vbox_text])
display(page)
```
