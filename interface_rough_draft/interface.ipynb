```python
import warnings
warnings.filterwarnings('ignore')

import helper_functions

import ipywidgets as widgets
from ipywidgets import *
from IPython.display import display, clear_output

from ipywidgets import Button, HBox, VBox
```

```python


file = open("paho_logo.jpg", "rb")
image= file.read()

image_headline = widgets.Image(
    value=image,
    format='jpg',
    width='200'
)

label_headline = widgets.Label(
    value='',
    style={'description_width': 'initial'}
)

vbox_headline = widgets.VBox([image_headline, label_headline])


Freight_ = widgets.Text(placeholder='(float64) Enter cost of freight charges for the shipment')

Country_ = widgets.Text(placeholder='(object) Enter the desitnation country receiving the shipment')

PurchaseOrderDate_ = widgets.DatePicker(description='Enter The date the purchase order was placed, in MM/DD/YYYY format')

Type_ = widgets.Text(placeholder='(object) Enter Product category, such as vaccine, syringe, or immunoglobulin.')

Product_ = widgets.Text(placeholder='(object) Enter 4-digit product code. The first two digits indicate the product tyupe, and the last two indicatE the verions')

Variant_ = widgets.Text(placeholder='(object) Enter 2-digit code  indicating dose count or presentation (e.g., vial or prefilled syringe)')

SupplierCountry_ = widgets.Text(placeholder='(object) Enter the origin country from which the product was supplied')

Quantity_ = widgets.Text(placeholder='(int64) Enter Number of unts included in the shipment')

Value_ = widgets.Text(placeholder='(float64) Enter total monetray value of the shipment')

#use api to handle AIR and BDI values

Doses_ = widgets.Text(placeholder='(int64) Enter Number of doses per vial')

#use backend to compute DISTANCE

Volume_ = widgets.Text(placeholder='(float64) Enter Total shipment volumn in cubic centimeters (cm^3)')

StorageCondition_ = widgets.IntSlider(
    value=0, #default = 0 degrees (idk if it's supposed to be fahrenheit or celsius)
    min = -50,
    max = 100,
    style={'description_width': 'initial', 'handle_color': '#16a085'}
)

#compute Openness in backend? as it relies on the receiving country that is an input

Weight_ = widgets.Text(placeholder='(float64) Enter weight per unit (vial, syringe, etc.) provided by javier, in kilograms')

TotalShipmentWeight_ = widgets.Text(placeholder='(float64) Enter weight of the entire shipment,calcued as Weight * Quantity / Doses, in kilograms')


widget_map = {
    "Freight_": Freight_,
    "Country_": Country_,
    "PurchaseOrderDate_": PurchaseOrderDate_,
    "Type_": Type_,
    "Product_": Product_,
    "Variant_": Variant_,
    "SupplierCountry_": SupplierCountry_,
    "Quantity_": Quantity_,
    "Value_": Value_,
    "Doses_": Doses_,
    "Volume_": Volume_,
    "StorageCondition_": StorageCondition_,
    "Weight_": Weight_,
    "TotalShipmentWeight_": TotalShipmentWeight_
}
```

```python
calculate_button = widgets.Button(
    description='Calculate',
    tooltip='Send',
    layout=Layout(width='50%', height='60px', grid_area='br'),
    style=ButtonStyle(button_color='lightblue'),
)

save_button  = Button(description='Save',
                  tooltip='Save',
                 layout=Layout(width='50%', height='60px', grid_area='br'),
                 style=ButtonStyle(button_color='lightblue'))


output = widgets.Output()

#add functionality to actually run calculations like API calls etc...
def on_button_clicked_calculate(event):
    with output:
        output.clear_output()
        
        html_content = "<h3>Inputted Variables:</h3><p>"

        for name, widget in widget_map.items():
            current_value = str(widget.value) 
            current_value = helper_functions.parse(current_value, name)
            
            if (widget.value == None) :
                current_value = ""
            html_content += f'<b>{name}</b>: {current_value} <br>'
        
        html_content += "</p>"
        
        html_text.value = html_content

        

#purely update the widget_map structure
def on_button_clicked_save(event):
    with output:
        output.clear_output()
        
        html_content = "<h3>Inputted Variables:</h3><p>"

        for name, widget in widget_map.items():
            current_value = str(widget.value) 
            if (widget.value == None) :
                current_value = ""
            html_content += f'<b>{name}</b>: {current_value} <br>'
        
        html_content += "</p>"
        
        html_text.value = html_content

        
       

calculate_button.on_click(on_button_clicked_calculate)
save_button.on_click(on_button_clicked_save)

vbox_result = widgets.VBox([calculate_button, save_button, output])

text_1 = widgets.HTML(value='<h3> Background API call to AIR & BDI values </h3>')
text_2 = widgets.HTML(value='<h3> DISTANCE will be computed in background. Here is temp output </h3>')
text_3 = widgets.HTML(value='<h3> idk what openness is.here is temp output </h3>')
text_4 = widgets.HTML(value='<h3> Enter the temperature shouldthe shipment be stored at </h3>')


html_text = widgets.HTML(value = "<p> this is out outputted </p>")
html_content = "<h3>Inputted Variables:</h3><p>"

for item in widget_map:
    html_content += f'<b>{item}</b>: <br>'

html_content += "</p>"
html_text.value = html_content

right_box = widgets.VBox([html_text])

vbox_text = widgets.VBox([Freight_, Country_, SupplierCountry_, Type_, Product_, Variant_, Quantity_, Value_, Doses_, Volume_, Weight_, TotalShipmentWeight_])
vbox_text2 = widgets.VBox([ text_1, text_2, text_4, StorageCondition_, text_3,  PurchaseOrderDate_, vbox_result])

page = widgets.HBox([vbox_headline, vbox_text, vbox_text2, right_box])

display(page)
```
